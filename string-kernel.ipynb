```python
import numpy as np
import pandas as pd 
import matplotlib
import collections, functools, operator 
```

```python
def w_phi(mistakes, x , p):
	result = 0
	if mistakes.shape[0]>0:
		for row in mistakes.iterrows():
			result += kernel(row[1][0], x, p)*row[1][1]
	return result
```

```python
def predict(x):
    if x<0:
        return -1
    else:
        return 1
```

```python
def mkernel(s, t, p):
	tlist = {}
	result = {}
	for i in range(len(t)-p+1):
		v = t[i:i+p]
		if v not in tlist:
			tlist[v] = 1

	for i in range(len(s)-p+1):
		v = s[i:i+p]
		if v not in tlist:
			result[v] = 0
		else:
			result[v] = 1
	result = sum(result.values())
	return result
```

```python
def kernel(s, t, p):
	tlist = {}
	result = {}
	for i in range(len(t)-p+1):
		v = t[i:i+p]
		if v not in tlist:
			tlist[v] = 1
		else:
			tlist[v] += 1

	for i in range(len(s)-p+1):
		v = s[i:i+p]
		if v not in tlist:
			tlist[v] = 0
		if v not in result:
			result[v] = tlist[v]
		else:
			result[v] += tlist[v]
	result = sum(result.values())
	return result
```

```python
def feature(s, y, p):
    result = {}
    for i in range(len(s)-p+1):
        v = s[i:i+p]
        if v not in result:
            result[v] = 1
        else:
            result[v] += 1
    if y<0: 
        result.update({n: -1 * result[n] for n in result.keys()})
    return result
```

```python
def kperceptron(df, p):
	mistakes = pd.DataFrame()
	for i in range(df.shape[0]):
		yt = df.iloc[i][1]
		xt = df.iloc[i][0]
		if yt*w_phi(mistakes, xt, p)<=0:
			mistakes = mistakes.append(df.iloc[i])
	return mistakes
```
